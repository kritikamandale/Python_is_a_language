```python
#Transaction Que

account_balance = 5000
max_transactions = 3
transaction_history = []

while True:
    transaction2 = int(input("Amount needs to be withdraw: "))
    if len(transaction_history) == max_transactions:
        print("Transaction failed: maximum transactions reached")
        break
    if account_balance - transaction2 < 0:
        print("insufficient balance")
        break
    account_balance -= transaction2
    transaction_history.append(transaction2)

    print("Remaining balance:", account_balance)
```

```python
def calculate_bill():
    cart_value = int(input("Enter cart value: "))
    num_products = int(input("Enter number of products: "))
    visits = int(input("Enter visits in last 2 months: "))

    history = []
    for i in range(visits):
        amount = int(input(f"Enter shopping amount for visit {i+1}: "))
        history.append(amount)

    discount = 0
    final_amount = cart_value

    # 20% discount
    if cart_value <= 3000 and num_products >= 10:
        discount = 0.20 * cart_value
        final_amount = cart_value - discount
        print("Discount Applied: 20%")

    # 30% discount
    elif cart_value > 3000 and cart_value <= 8000:
        discount = 0.30 * cart_value
        final_amount = cart_value - discount
        print("Discount Applied: 30%")

    # 50% discount
    elif cart_value > 8000:
        count = 0
        for amt in history:
            if amt >= 20000:
                count += 1

        if visits >= 4 and count >= 4:
            discount = 0.50 * cart_value
            final_amount = cart_value - discount
            print("Discount Applied: 50%")
        else:
            print("50% Discount not applicable")

    else:
        print("No Discount Applied")

    print("Final Amount to Pay:", final_amount)


# Call function
calculate_bill()
```
