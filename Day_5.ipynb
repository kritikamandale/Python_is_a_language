```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")

print("Original Shape : ",df.shape)
print("\nFirst 5 rows:")
print(df.head(10))
```

```python
#Age distribution

plt.figure()
sns.histplot(df['age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
#fare distribution

plt.figure()
sns.histplot(df['fare'],bins=20, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
#boxplot (fare by class)
plt.figure()
sns.boxplot(x='class', y='fare', data=df)
plt.title("Fare by Class")
```

```python
#Violin plot (Age by class)
plt.figure()
sns.violinplot(x='class', y='age', data=df)
plt.title("Age distribution by Class")
plt.show()
```

```python
#correlation Heatmap

plt.figure()
corr = df.corr(numeric_only= True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("spotify-tracks-dataset.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.describe()
df.isnull().sum()
```

```python
#1. Distribution of Popularity (Histogram)
plt.figure()
plt.hist(df['popularity'], bins=20)
plt.title("Distribution of Track Popularity")
plt.xlabel("Popularity")
plt.ylabel("Count")
plt.show()
```

```python
#2. Top 10 Most Popular Genres (Bar Chart)
top_genres = df['track_genre'].value_counts().head(10)

plt.figure()
top_genres.plot(kind='bar')
plt.title("Top 10 Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Tracks")
plt.show()
```

```python
#3. Energy vs Popularity (Scatter Plot)
plt.figure()
plt.scatter(df['energy'], df['popularity'])
plt.title("Energy vs Popularity")
plt.xlabel("Energy")
plt.ylabel("Popularity")
plt.show()
```

```python
#4. Danceability Distribution
plt.figure()
plt.hist(df['danceability'], bins=20)
plt.title("Danceability Distribution")
plt.show()
```

```python
#5. Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#6. Tempo Distribution
plt.figure()
plt.hist(df['tempo'], bins=30)
plt.title("Tempo Distribution")
plt.xlabel("Tempo")
plt.show()
```

```python
#7. Duration Distribution (converted to minutes)
df['duration_min'] = df['duration_ms'] / 60000

plt.figure()
plt.hist(df['duration_min'], bins=30)
plt.title("Track Duration (Minutes)")
plt.xlabel("Duration (min)")
plt.show()
```

```python
#8. Get top 5 genres
top5 = df['track_genre'].value_counts().head(5).index
df_top5 = df[df['track_genre'].isin(top5)]

plt.figure(figsize=(10,6))
sns.violinplot(x='track_genre', y='energy', data=df_top5)

plt.title("Energy Distribution by Top 5 Genres")
plt.xticks(rotation=45)
plt.show()
```
